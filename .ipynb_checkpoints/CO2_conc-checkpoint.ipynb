```python
%matplotlib notebook
import random
random.seed()
import matplotlib.pyplot as plt
import numpy as np
from collections import namedtuple
from typing import Union, List
from scipy.optimize import least_squares


K_h = 1 / 29.41
P_CO2_range = np.linspace(0.2, 46, 200)
K1 = 10**-6.4
K2 = 10**-10.3
pH = []

H2CO3 = []
HCO3 = []
CO3 = []
pH_range = np.linspace(3, 14, 200)

for p_co2 in P_CO2_range:
    poly = np.poly1d([1, 0, -1*(10**-14 + K1 * K_h * p_co2), -2 * K1 * K2 * K_h * p_co2])
    pH.append(np.log10(poly.r[0].real))

    
p = 10
ct = p * K_h
for pH_i in pH_range:
    Hplus = 10**(-1 * pH_i)
    z = Hplus**2 + Hplus * K1 + K1 * K2
    CO3.append(K1 * K2 / z * ct)
    HCO3.append(Hplus * K1 / z * ct)
    H2CO3.append(Hplus**2 / z * ct)

CO3 = np.array(CO3)    
HCO3 = np.array(HCO3)  
H2CO3 = np.array(H2CO3)

fig1, ax1 = plt.subplots() 
ax1.plot(P_CO2_range, pH)
ax1.set_xlabel('Pressure CO2, bar')
ax1.set_ylabel('pH')
ax1.set_title('pH = f(P_CO2)')
ax1.grid(True)

fig2, ax2 = plt.subplots() 
ax2.plot(pH_range, CO3 / ct, 'b')
ax2.plot(pH_range, HCO3 / ct, 'p')
ax2.plot(pH_range, H2CO3 / ct, 'r')
ax2.set_ylabel('alpha (HxCO3/Ct)')
ax2.set_xlabel('pH')


fig3, ax3 = plt.subplots() 
ax3.plot(pH_range, np.log10(CO3 / ct), 'b')
ax3.plot(pH_range, np.log10(HCO3 / ct), 'p')
ax3.plot(pH_range, np.log10(H2CO3 / ct), 'r')
ax2.set_ylabel('alpha (HxCO3/Ct)')
ax3.set_xlabel('pH')

plt.show()



```
